```python
# Free cell — no license needed
import math

data = [1, 4, 9, 16, 25, 36, 49, 64, 81, 100]
sqrt_data = [math.sqrt(x) for x in data]
mean = sum(sqrt_data) / len(sqrt_data)

print(f'Data: {data}')
print(f'Square roots: {[round(x, 2) for x in sqrt_data]}')
print(f'Mean of sqrt: {mean:.2f}')
```

```python
# Another free cell
def fibonacci(n):
    a, b = 0, 1
    for _ in range(n):
        yield a
        a, b = b, a + b

fibs = list(fibonacci(15))
print('Fibonacci sequence:', fibs)
```
