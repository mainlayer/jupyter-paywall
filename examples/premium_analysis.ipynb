````python
# Free cell — runs without a license
import random

# Generate some sample sales data
products = ['Widget A', 'Widget B', 'Widget C', 'Widget D']
sales_data = {
    product: [random.randint(100, 1000) for _ in range(12)]
    for product in products
}

print('Sample data generated.')
for product, sales in sales_data.items():
    print(f'  {product}: total={sum(sales)}, avg={sum(sales)/12:.0f}')
````

````python
%%mainlayer_paid
# PREMIUM CELL: Advanced statistical analysis
# Requires a Mainlayer license

import statistics

print('=== Premium Statistical Analysis ===')
for product, monthly_sales in sales_data.items():
    mean = statistics.mean(monthly_sales)
    std = statistics.stdev(monthly_sales)
    median = statistics.median(monthly_sales)
    trend = 'up' if monthly_sales[-1] > monthly_sales[0] else 'down'
    
    print(f'\n{product}:')
    print(f'  Mean:   {mean:.1f}')
    print(f'  Std:    {std:.1f}')
    print(f'  Median: {median}')
    print(f'  Trend:  {trend}')
````

````python
%%mainlayer_paid
# PREMIUM CELL: Forecasting
# Requires a Mainlayer license

def simple_moving_average(data, window=3):
    return [
        sum(data[i:i+window]) / window
        for i in range(len(data) - window + 1)
    ]

def naive_forecast(data, periods=3):
    """Naive forecast using last period + average trend."""
    if len(data) < 2:
        return [data[-1]] * periods
    trend = (data[-1] - data[0]) / (len(data) - 1)
    return [round(data[-1] + trend * (i + 1)) for i in range(periods)]

print('=== Premium Sales Forecast (Next 3 months) ===')
for product, monthly_sales in sales_data.items():
    sma = simple_moving_average(monthly_sales)
    forecast = naive_forecast(monthly_sales, 3)
    print(f'{product}: {forecast}')
````
